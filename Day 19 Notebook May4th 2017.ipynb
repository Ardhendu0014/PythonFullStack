```python
def count_em_up(phrase):    
    word_counts = dict()                # Initialize Empty Dict

    clean_phrase = phrase.lower()
    split_ups = clean_phrase.split()    # A list of words

    for word in split_ups:
        word_counts[word] = word_counts.get(word, 0) + 1
    
    return word_counts
```
