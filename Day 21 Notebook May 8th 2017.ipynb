```python
class LightningTalisman:
    """This is a brief summary. Follow PEP 257."""
    
    def __init__(self, divine_power, flashes=False):
        """
        By convention, this is first.
        
        In many languages known as a "Constructor".
        Really an "initializer" - A set of instructions to make 
        a single instance of this type.
        """
        self.divine_power = divine_power
        self.flashes = flashes

    def __repr__(self):
        """
        The Representation.
        
        A way to "look" at one LightningTalisman.
        
        """
        return "A lightning talisman."
    
    def smiting(self):
        """An instance method"""
        print('Smites You.')
        
        
```

```python
vajra = LightningTalisman(divine_power='enlightenment') # Instantiation
```

```python
print(vajra)        # prints out the __repr__
```

```python
vajra.smiting()   # Invoking an instance method
```

```python
vajra.divine_power    # Access by attribute
```

```python
class Vehicle:
    """
    This is an abstract of the objects purpose.
    """
    
    def __init__(self, color, doors=4, tires=4, wheels=4, cylinders=4):
        self.color = color
        self.doors = doors
        self.tires = tires
        self.wheels = wheels
        self.cylinders = cylinders
    
    def turn(self, direction: str) -> None:
        """Instance Method"""
        message = "Makes a {} turn".format(direction)
        return message

```

```python
ferrari = Vehicle(color='red')
```

```python
ferrari.color    # Access by attribute
```

```python
ferrari.wheels
```

```python
captured = ferrari.turn('right')
```

```python
print(captured)
```

```python
class Monument:
    def __init__(self, location, title, material, age, memo):
        self.location = location
        self.title = title
        self.material = material
        self.age = age
        self.memo = memo
        
    def __repr__(self):
        return "Monument: {}.".format(self.title)
```

```python
def collect_data():

    yellowstone_arch = Monument('Wyoming', 'Yellowstone National Park',
                                'earth', 100, 'Named in 1872')
    
    crates_of_the_moon = Monument('Idaho', 'Craters of the Moon Natl Monument.',
                             'Volvanic Rock', 10000000, 'From the moon.')
    
    
    monuments = [yellowstone_arch, crates_of_the_moon]
    result = [m for m in monuments if m.age > 100]
    
    return result
```

```python
collect_data()
```
