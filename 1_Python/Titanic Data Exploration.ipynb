```python
from sklearn import datasets
from sklearn import metrics
from sklearn.ensemble import ExtraTreesClassifier
from matplotlib import pyplot as plt
from collections import Counter
import pandas
import re

```

```python
df = pandas.read_csv('/home/k/Desktop/titanic3.csv')
```

```python
ages = list(df['age'])
sex = list(df['sex'])
```

```python
%matpotlib inline


ac = Counter(ages)
sc = Counter(sex)

common = sc.most_common()
sexes, scores = list(zip(*common))

fig1, ax1 = plt.subplots()
ax1.pie(scores, labels=sexes, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# plt.show()
```

```python
def sex_extractor(sex):
    if 'f' in sex:
        return 1
    elif 'm' in sex:
        return 0
    
    
def deck_extractor(cabin):
    
    decks = {l.upper(): n for n, l in enumerate('abcdefgt')}
    deck = re.sub('[^A-Z]', '', cabin)[0]
    
    return decks[deck]
```

```python
df['sex'] = df['sex'].apply(sex_extractor)


df = df.drop(df.columns[[2, 7, 10, 11, 12, 13]], axis=1)
df = df.dropna()

df['deck'] = df['cabin'].apply(deck_extractor)

target = list(df['survived'])
df = df.drop('survived', 1)
df = df.drop('cabin', 1)
```

```python
t_deck = Counter(sorted(df['deck']))
list(sorted(t_deck.most_common()))
```

```python
model = ExtraTreesClassifier()     # Fit an Extra Trees model to the data.

model.fit(df, target)             

print(model.feature_importances_)  # Display the relative importance of each attribute.
```
